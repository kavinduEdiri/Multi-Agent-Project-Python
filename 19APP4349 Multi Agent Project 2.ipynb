```python
class StudentRegistrationAgent:
    def __init__(self):
        self.student_info = {}

    def collect_student_details(self):
        print("Student Registration Agent: Please provide your details.")
        self.student_info['name'] = input("Enter your name: ")
        self.student_info['student_id'] = input("Enter your student ID: ")
        print(f"Student Registration Agent: Welcome, {self.student_info['name']} (ID: {self.student_info['student_id']}). Proceeding to course selection.")
        return self.student_info


class CourseSelectionAgent:
    def __init__(self):
        self.courses = {
            "Data Structures": 300,
            "Algorithms": 350,
            "Machine Learning": 400
        }

    def display_available_courses(self):
        print("Course Selection Agent: The following courses are available:")
        for idx, (course, price) in enumerate(self.courses.items(), 1):
            print(f"{idx}. {course} - ${price}")

    def select_courses(self):
        self.display_available_courses()
        print("Please type the names of the courses you wish to enroll in, separated by commas.")
        selected_courses = input("Enter your selected courses: ").split(",")
        selected_courses = [course.strip() for course in selected_courses]
        total_cost = sum(self.courses.get(course, 0) for course in selected_courses)
        print(f"Course Selection Agent: You have chosen: {', '.join(selected_courses)}. Total cost: ${total_cost}.")
        return selected_courses, total_cost


class PaymentAgent:
    def handle_payment(self, amount_due):
        print(f"Payment Agent: Your total payment due is ${amount_due:.2f}.")
        payment_received = float(input("Enter the payment amount: "))
        if payment_received >= amount_due:
            print(f"Payment Agent: Payment of ${payment_received:.2f} received. Registration completed successfully!")
            return True
        else:
            print(f"Payment Agent: Payment of ${payment_received:.2f} is insufficient. Registration failed.")
            return False



if __name__ == "__main__":
    print("Student: I want to register for courses.")

    
    student_agent = StudentRegistrationAgent()
    course_agent = CourseSelectionAgent()
    payment_agent = PaymentAgent()

   
    student_details = student_agent.collect_student_details()

    
    chosen_courses, total_fee = course_agent.select_courses()

    
    payment_success = payment_agent.handle_payment(total_fee)

    
    if payment_success:
        print(f"Student: Thank you, {student_details['name']}! Your registration is now complete.")
    else:
        print(f"Student: Registration incomplete. Please try again.")
```
